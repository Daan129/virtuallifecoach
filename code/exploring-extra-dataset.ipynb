```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import DateTime as dt
```

```python
df = pd.read_csv('extra-dataset.csv')
df.head(50)
```

```python
df['average mood'] = df[['mood breakfast','mood lunch', 'mood diner']].mean(axis=1)
df.head()
```

```python
av_by_day = df.groupby("day")["average mood"].mean()
av_by_day
```

```python
av_by_user = df.groupby("user id")["average mood"].mean()
av_by_user
```

```python
av_by_stuwo = df.groupby("study/work day")["average mood"].mean()
av_by_stuwo
```

```python
df_geenid = df[['study/work day','h/o sleep','h/o exercise','h/o relaxation','h/o study','h/o work','h/o other', 'average mood']]
df_geenid
```

```python
sns.pairplot(df)
plt.show()
```

```python
df['h/o sleep'] = pd.to_datetime(df['h/o sleep'], format='%H:%M:%S')
df['h/o exercise'] = pd.to_datetime(df['h/o exercise'], format='%H:%M:%S')
df['h/o relaxation'] = pd.to_datetime(df['h/o relaxation'], format='%H:%M:%S')
df['h/o study'] = pd.to_datetime(df['h/o study'], format='%H:%M:%S')
df['h/o work'] = pd.to_datetime(df['h/o work'], format='%H:%M:%S')
df['h/o other'] = pd.to_datetime(df['h/o other'], format='%H:%M:%S')
```

```python
df.dtypes
```

```python
df['sleep'] = (df['h/o sleep'].dt.hour + (df['h/o sleep'].dt.minute / 60))
df['exercise'] = (df['h/o exercise'].dt.hour + (df['h/o exercise'].dt.minute / 60))
df['relaxation'] = (df['h/o relaxation'].dt.hour + (df['h/o relaxation'].dt.minute / 60))
df['study'] = (df['h/o study'].dt.hour + (df['h/o study'].dt.minute / 60))
df['work'] = (df['h/o work'].dt.hour + (df['h/o work'].dt.minute / 60))
df['other'] = (df['h/o other'].dt.hour + (df['h/o other'].dt.minute / 60))
```

```python
df_sub = df[['study/work day','sleep','exercise','relaxation','study','work','other','mood breakfast','mood lunch','mood diner', 'average mood']]
```

```python
sns.pairplot(df_sub)
plt.show()
```
